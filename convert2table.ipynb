```
file_loc = "gdrive/My Drive/Colab/Shakespeare Emotion Analyses/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt"
file1 = open(file_loc, 'r') 
lines = file1.readlines() 
```

```
file_loc = "gdrive/My Drive/Colab/Shakespeare Emotion Analyses/Emotion-Lexicon.csv"
text_file = open(file_loc, "w")
```

```
check_first = True
previous_word = ""
for line in lines[1:]:
  informations = line.strip().split("\t")
  word = informations[0]
  emotion = informations[1]
  value = informations[2]
  if check_first:
    n = text_file.write("words\t")
    n = text_file.write("anger\t")
    n = text_file.write("anticipation\t")
    n = text_file.write("disgust\t")
    n = text_file.write("fear\t")
    n = text_file.write("joy\t")
    n = text_file.write("negative\t")
    n = text_file.write("positive\t")
    n = text_file.write("sadness\t")
    n = text_file.write("surprise\t")
    n = text_file.write("trust\t")
    check_first = False
  if previous_word != word:
    n = text_file.write("\n")
    n = text_file.write(word+"\t")
    n = text_file.write(value+"\t")
    previous_word = word
  else:
    n = text_file.write(value+"\t")
```

```
text_file.close()
```
